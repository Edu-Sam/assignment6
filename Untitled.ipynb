```python
#import plotly.graph_objects as go
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd 
import numpy as np
from plotly.offline import init_notebook_mode, iplot
#fig = go.Figure(data=go.Bar(y=[2, 3, 1]))
#fig.show()
```

```python
titanic_data=pd.read_csv('Titanic Data.csv');
print(titanic_data);
```

```python

fig = go.Figure(data=[go.Histogram(x=titanic_data["Age"], histnorm='percent')])

fig.show()
```

```python
titanic_median=titanic_data["Age"].median()
print(titanic_median)
```

```python
titanic_data["Age"].fillna(titanic_median, inplace = True) 
print(titanic_data)
```

```python
fig1=go.Figure()
male=titanic_data[titanic_data["Sex"]=="male"]
female=titanic_data[titanic_data["Sex"]=="female"]
fig1.add_trace(go.Histogram(x=male,histnorm='percent',
    name='male'))
fig1.add_trace(go.Histogram(x=female,histnorm='percent',
    name='female'))

# Overlay both histograms
fig1.update_layout(barmode='overlay')
# Reduce opacity to see both histograms
fig1.update_traces(opacity=0.75)
fig1.show()
```
